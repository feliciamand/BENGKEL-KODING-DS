```python
import matplotlib.pyplot as plt
import seaborn as sns

# Informasi umum dataset
info = df.info()

# Deskripsi statistik fitur numerik
desc = df.describe()

# Cek missing values
missing = df.isnull().sum()

# Cek duplikasi
duplicates = df.duplicated().sum()

# Cek nilai unik untuk setiap kolom
unique_vals = df.nunique()

# Distribusi target
target_dist = df['NObeyesdad'].value_counts()

# Visualisasi distribusi target
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='NObeyesdad', order=df['NObeyesdad'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Distribusi Kategori NObeyesdad (Obesity Level)')
plt.tight_layout()
plt.show()

# Deteksi outlier dengan boxplot untuk fitur numerik
numerical_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

# Tampilkan hasil yang penting ke pengguna
import ace_tools as tools

tools.display_dataframe_to_user(name="Statistik Data Numerik", dataframe=desc)
```
